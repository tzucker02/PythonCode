```python
from tabulate import tabulate

data = [["Banana", "$2"], ["Apple", "$3"]]
headers = ["Fruit", "Price"]

print(tabulate(data, headers=headers, tablefmt="fancy_grid"))
```

```python
from prettytable import PrettyTable
t = PrettyTable(["Fruit", "Price"])
[t.add_row(r) for r in [["Banana", "$2"], ["Apple", "$3"]]]

print(t)
```

```python
from terminaltables import AsciiTable

table_data = [    ["Fruit", "Price"],
    ["Banana", "$2"],
    ["Apple", "$3"]
]

table = AsciiTable(table_data)
print(table.table)
```

```python
from rich.console import Console
from rich.table import Table

t = Table()
[t.add_column(c) for c in ["Fruit", "Price"]]
[t.add_row(*r) for r in [["Banana", "$2"], ["Apple", "$3"]]]

Console().print(t)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from pandas.plotting import table

# Your data
df = pd.DataFrame([["Banana", "$2"], ["Apple", "$3"]], columns=["Fruit", "Price"])

# Plot setup
fig, ax = plt.subplots()
ax.axis('off')  # Hide axes
table(ax, df, loc='center')

plt.show()
```

```python
import plotly.graph_objects as go

fig = go.Figure(data=[
    go.Table(
        header=dict(values=["Fruit", "Price"]),
        cells=dict(values=[["Banana", "Apple"], ["$2", "$3"]])
    )
])
fig.show()
```

```python
from pytablewriter import MarkdownTableWriter

w = MarkdownTableWriter(
    headers=["Fruit", "Price"],
    value_matrix=[["Banana", "$2"], ["Apple", "$3"]]
)

print(w.dumps())
```
